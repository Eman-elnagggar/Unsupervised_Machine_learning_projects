# Unsupervised Model

```python
Customer_data_category=Customer_Segmentation_data.select_dtypes(include=['category'])
Customer_data_number=Customer_Segmentation_data.select_dtypes(exclude=['category'])
```

```python
Customer_data_category.info()
```

```python
Customer_data_number.info()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
for col in Customer_data_category:
  Customer_data_category[col]=encoder.fit_transform(Customer_data_category[col])
```

```python
Customer_data_category.info()
```

```python
x=pd.concat([Customer_data_category,Customer_data_number],axis=1)
```

```python
x=x.reset_index(drop=True)
```

```python
x.info()
```

```python
x.isnull().sum()
```

```python
from sklearn.preprocessing import MinMaxScaler
min_max_scaler=MinMaxScaler()
x=min_max_scaler.fit_transform(x)
```

```python
from sklearn.cluster import KMeans
inertia = []
for k in range(1,11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(x)
    inertia.append(kmeans.inertia_)


plt.plot(range(1,11), inertia, 'bo-')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method For Optimal k')
plt.show()
```

- I think 5 is good

```python
kmeans=KMeans(n_clusters=5,random_state=42)
kmeans.fit_transform(x)
kmean_prdict=kmeans.predict(x)
```

```python
kmean_prdict
```

# supervised learning

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,kmean_prdict,test_size=0.2,random_state=42,shuffle=True)
```

- logistic regression can not fit because it is not a binary classification

- try decision_tree

## Decision Tree

- I choose decision tree cause data is small and this algorithm has power to reduce overfitting by spiliting data into small subsets

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc=DecisionTreeClassifier(criterion="gini", max_depth=3)
decision_model=dtc.fit(x_train,y_train)
decision_prediction=decision_model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, decision_prediction)
print(f"Accuracy: {accuracy:.2f}")
```
